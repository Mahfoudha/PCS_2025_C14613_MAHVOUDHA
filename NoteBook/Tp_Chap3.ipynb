## Exercise 5

```python
import math
import matplotlib.pyplot as plt

h = 0.001
a = -0.5
u = [math.exp(0), math.exp(a), math.exp(2*a)]
n_max = 1000

for n in range(3, n_max+1):
    next_u = u[-1] + h * (23/12*u[-1] - 4/3*u[-2] + 5/12*u[-3])
    u.append(next_u)

t = [n*h for n in range(n_max+1)]
exact = [math.exp(a*n*h) for n in range(n_max+1)]

plt.plot(t, u, label='Numerical')
plt.plot(t, exact, label='Exact', linestyle='dashed')
plt.title('Numerical vs Exact Solution')
plt.xlabel('t')
plt.ylabel('u')
plt.legend()
plt.grid(True)
plt.show()

error = [abs(u_val - e_val) for u_val, e_val in zip(u, exact)]

plt.plot(t, error)
plt.title('Error between Numerical and Exact Solution')
plt.xlabel('t')
plt.ylabel('Error')
plt.grid(True)
plt.show()
```

## Exercise 6

```python
def symmetric_difference(A, B):
    return (A - B) | (B - A)

A = {1, 2, 3}
B = {3, 4, 5}
print("Custom symmetric difference:", symmetric_difference(A, B))
print("Built-in symmetric difference:", A.symmetric_difference(B))
```
